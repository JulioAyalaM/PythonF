```python
# Crea una gráfica de barras que represente a los 10 países con más medallas ganadas


import pandas as pd
import numpy as np

#se lee el archivo fuente
datos1=pd.read_csv("C:\\Users\\juayala\OneDrive - Sigma\\Downloads\\athlete_events.csv")
datos1
```

```python
#Excluyo los datos de los paises que no tienen medalla NAN
jugadoresMedallas = datos1[datos1["Medal"].notna()]
```

```python
# obtengo datos de medallas por pais - y se ordena de forma descendente para tomar un top 10. 
MedallasxPais = jugadoresMedallas.groupby(['NOC'],as_index=False).Medal.count().sort_values(by='Medal',ascending=False).head(10)
MedallasxPais
```

```python
#Crear la grafica
MedallasxPais.set_index('NOC',inplace=True)
MedallasxPais["Medal"].plot(kind='bar')
```
