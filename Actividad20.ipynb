# Crea / Presenta un algoritmo que genere los datos de "reading_score" y "math_score", en variables categóricas, y guárdalo en dos columnas diferentes (cada columna nueva representa la nueva columna con variables categórica).


```python
import pandas as pd
import numpy as np
import matplotlib as plt

#se lee el archivo fuente
data_students=pd.read_csv("C:\\Users\\juayala\\Documents\\CursoPython\\clean_students_complete.csv")
data_students
```

```python
#Utilizaremos la funcion cut del dataframe para hacer la conversion a variables categóricas

data_students['reading_score_cat'] = pd.cut(data_students.reading_score, bins =[0,69,float('Inf')] , labels=['Reprobado','Aprobado'])
data_students['math_score_cat'] = pd.cut(data_students.reading_score, bins =[0,69,float('Inf')] , labels=['Reprobado','Aprobado'])


data_students
```

# Crea / Presenta una gráfica que condense la información obtenida ahora categóricamente

```python
totalesAprobadosReading = data_students.value_counts("reading_score_cat")
etiquetas = data_students["reading_score_cat"].value_counts().keys()

graficaAprobadosReading = totalesAprobadosReading.plot.pie(figsize=(10,10) ,labels=etiquetas,autopct="%0.5f" )
```

```python
totalesAprobadosMath = data_students.value_counts("math_score_cat")
etiquetas = data_students["math_score_cat"].value_counts().keys()

graficaAprobadosMath = totalesAprobadosMath.plot.pie(figsize=(10,10) ,labels=etiquetas,autopct="%0.5f" )
```
