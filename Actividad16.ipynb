```python
import pandas as pd
import numpy as np

datos1=pd.read_csv("C:\\Users\\juayala\OneDrive - Sigma\\Downloads\\athlete_events.csv")
datos1
```

```python
# primero hay que obtener el valor del año mas reciente (Max)
MaxYear = datos1["Year"].max()
MaxYear
```

```python
datos2016 = datos1[datos1["Year"] == 2016 ]
datos2016
```

```python
#Excluir valores nan para quedarme unicamente con los registros que tienen medalla
datosMedallas2016 = datos2016[datos2016["Medal"].notna()]
datosMedallas2016
```

```python
# distribucion de medallas por paises
# MedallasxPais = datosMedallas2016.groupby(['NOC'],as_index=False).Medal.count().sort_values(by='Medal',ascending=False).head()
# MedallasxPais

datosIzquierda = datosMedallas2016[["Medal"]]
datosIzquierda
```

```python
dummies = pd.get_dummies(datosIzquierda[["Medal"]])
dummies
```

```python
concat = pd.concat([datosIzquierda, dummies ] , axis = 1)
concat
```

```python
agrupado = concat.groupby(["Medal"]).sum()
agrupado
```

```python
agrupado["Totals"] = agrupado["Medal_Bronze"] + agrupado["Medal_Gold"] + agrupado["Medal_Silver"]
agrupado
```

```python
# ahora si vamos a crear la grafica
etiquetas = datosMedallas2016["Medal"].value_counts().keys()

pie = agrupado["Totals"].plot.pie(figsize=(10,10) ,labels=etiquetas,autopct="%0.5f" )
```

```python
# Guardar la grafica
figura = pie.get_figure()
figura.savefig("Grafica2016.png")
```
