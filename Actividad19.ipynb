# Crea / Presenta la forma que creas más conveniente para describir qué grado por escuela tuvo el mejor puntaje en "reading" y con ello saber que genero es el que predomina en este filtro.

```python
import pandas as pd
import numpy as np
import matplotlib as plt

#se lee el archivo fuente
data_students=pd.read_csv("C:\\Users\\juayala\\Documents\\CursoPython\\clean_students_complete.csv")
data_students
```

```python
# primero necesitamos obtener cual fue el mejor puntaje por grado y escuela
mejoresPuntajes = data_students.groupby(['school_name', 'grade'],as_index=False).reading_score.max().sort_values(by='reading_score',ascending=False).head(20)
mejoresPuntajes
```

```python
# De acuerdo a la informacion anterior, el mayor puntaje de cada escuela fue de 99 
# Entonces procederemos a buscar cuantos alumnos en cada escuela obtuvieron dicha puntuacion

mejoresPuntajesxEscuela =data_students[data_students["reading_score"] == 99 ].groupby(["school_name","grade"]).agg({'Student ID': 'count', 'reading_score':'max'})
mejoresPuntajesxEscuela
```

```python
#Para saber que genero predomina cambiaremos la agrupacion agregando el genero

mejoresPuntajesxEscuelaGenero = data_students[data_students["reading_score"] == 99 ].groupby(["school_name","grade","gender"]).agg({'Student ID': 'count', 'reading_score':'max'})

with pd.option_context('display.max_rows', None,'display.max_columns', None,'display.precision', 3,):
    print(mejoresPuntajesxEscuelaGenero)
```

```python
#Debido a que el dataframe es algo grande para mostrar podriamos guardarlo en csv para analizarlo mas a detalle
mejoresPuntajesxEscuelaGenero.to_csv("informacionXgrado.csv", header = True,index = True )
```
