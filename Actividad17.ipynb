# Crea/Presenta  la forma que creas más conveniente para describir cuántos alumnos hay por grados.

```python
import pandas as pd
import numpy as np
import matplotlib as plt

#se lee el archivo fuente
data_students=pd.read_csv("C:\\Users\\juayala\\Documents\\CursoPython\\clean_students_complete.csv")
data_students
```

```python
# La forma mas rapida para hacerlo creo que seria con value_count, sin mas complicacion

totales = data_students.value_counts("grade")
totales
```

```python
# graficamos los resultados
etiquetas = data_students["grade"].value_counts().keys()
grafica = totales.plot.pie(figsize=(10,10) ,labels=etiquetas,autopct="%0.5f" )
```
