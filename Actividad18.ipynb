# Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el mejor puntaje en "reading", y con ello saber que género es el que predomina en este filtro.

```python

import pandas as pd
import numpy as np
import matplotlib as plt

#se lee el archivo fuente
data_students=pd.read_csv("C:\\Users\\juayala\\Documents\\CursoPython\\clean_students_complete.csv")
data_students
```

```python
# primero necesitamos obtener cual fue el mejor puntaje por escuela
mejoresPuntajes = data_students.groupby(['school_name'],as_index=False).reading_score.max().sort_values(by='reading_score',ascending=False).head(20)
mejoresPuntajes
```

```python
# De acuerdo a la informacion anterior, el mayor puntaje de cada escuela fue de 99 
# Entonces procederemos a buscar cuantos alumnos en cada escuela obtuvieron dicha puntuacion

mejoresPuntajesxEscuela =data_students[data_students["reading_score"] == 99 ].groupby(["school_name"]).agg({'Student ID': 'count', 'reading_score':'max'})
mejoresPuntajesxEscuela

```

```python
#Para saber que genero predomina cambiaremos la agrupacion agregando el genero
mejoresPuntajesxEscuelaGenero = data_students[data_students["reading_score"] == 99 ].groupby(["school_name","gender"]).agg({'Student ID': 'count', 'reading_score':'max'})
mejoresPuntajesxEscuelaGenero
```

```python
#Ahora intentaremos representarlo con un método diferente, usando la matriz identidad 
filtro = data_students[data_students["reading_score"] == 99 ]
filtro

datosIzquierda =     filtro[["school_name"]]
datosIzquierda
```

```python
dummies = pd.get_dummies(filtro[["gender"]])
dummies
```

```python
concat = pd.concat([datosIzquierda, dummies ] , axis = 1)
concat
```

```python
agrupados = concat.groupby(["school_name"]).sum()
agrupados
```

```python
agrupados["Predominante"] = "None"

for index, row in agrupados.iterrows():
    
    if row["gender_F"] > row["gender_M"]:
        resultado = "Femenimo"
    elif row["gender_F"] < row["gender_M"]:
        resultado = "Masculino"
    else:
        resultado = "Iguales"
    
    agrupados.at[index,'Predominante'] = resultado

agrupados
```

# Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el peor puntaje en "reading", y con ello saber que género es el que predomina en este filtro.

```python
PeoresPuntajes = data_students.groupby(['school_name'],as_index=False).reading_score.min().sort_values(by='reading_score',ascending=False).head(20)
PeoresPuntajes
```

```python
#Con lo anterior sabemos que los peores puntajes fueron de 69 en algunas escuelas y en el resto de 63

PeoresPuntajesxEscuela69 = data_students[data_students["reading_score"] == 69].groupby(["school_name"]).agg({'Student ID': 'count', 'reading_score':'min'})
PeoresPuntajesxEscuela63 = data_students[data_students["reading_score"] == 63].groupby(["school_name"]).agg({'Student ID': 'count', 'reading_score':'min'})

PeoresPuntajesxEscuela = pd.concat([PeoresPuntajesxEscuela69, PeoresPuntajesxEscuela63])
PeoresPuntajesxEscuela
```

```python
#Para saber que genero predomina cambiaremos la agrupacion agregando el genero
PeoresPuntajesxEscuelaGenero69 = data_students[data_students["reading_score"] == 69 ].groupby(["school_name","gender"]).agg({'Student ID': 'count', 'reading_score':'min'})
PeoresPuntajesxEscuelaGenero63 = data_students[data_students["reading_score"] == 63 ].groupby(["school_name","gender"]).agg({'Student ID': 'count', 'reading_score':'min'})

PeoresPuntajesxEscuelaGenero = pd.concat([PeoresPuntajesxEscuelaGenero69, PeoresPuntajesxEscuelaGenero63])
PeoresPuntajesxEscuelaGenero
```
